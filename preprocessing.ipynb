```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data_path = "C:/Users/PC/Desktop/IKCO_preprocessing/Stock_IKCO1.csv"
data = pd.read_csv(data_path)
data.head()
```

```python
df = data.copy()
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
df = df.rename(columns = {"<TICKER>" : "ticker", 
                          "<DTYYYYMMDD>" : "date", 
                          "<OPEN>" : "open", 
                          "<HIGH>" : "high", 
                          "<LOW>" : "low", 
                          "<CLOSE>" : "close",
                          "<VALUE>" : "value",
                          "<OPENINT>" : "openint",
                          "<PER>" : "per",
                          "<LAST>" : "last",
                          "<VOL>" : "volume"})
df.head()
```

```python
df["date"] = pd.to_datetime(df["date"].astype(str), format="%Y%m%d")
df.head()
```

```python
df = df.set_index("date").sort_index()
df.head()
```

```python
df.info()
df.head()
```

```python
dup_dates = df.index.duplicated().sum()
dup_dates
```

```python
mask_valid = ((df["open"]>0) & (df["high"]>0) & (df["low"]>0) & (df["close"]>0) & (df["volume"]>0))
df = df[mask_valid]
df.describe()
```

```python
df["return"] = df["close"].pct_change()
df.head()
```

```python
df["log_return"] = np.log(df["close"] / df["close"].shift(1))
df.head()
```

```python
df[["close", "return", "log_return"]].head(10)
```

```python
df['ma_5'] = df["close"].rolling(window=5).mean()
df[["close", "ma_5"]].head(10)
```

```python
df["ma_20"] = df["close"].rolling(window=20).mean()

df[["close", "ma_5", "ma_20"]].head(25)
```

```python
df["vol_20"] = df["return"].rolling(window=20).std()

df[["return", "vol_20"]].head(25)
```

```python
plt.figure(figsize=(10, 4))
df["close"].plot(label="close", alpha=0.6)
df["ma_5"].plot(label="ma_5", alpha=0.8)
df["ma_20"].plot(label="ma_20", alpha=0.8)
plt.legend()
plt.title("Close with 5D and 20D Moving Averages")
plt.show()
```

```python
df.isna().sum()
```

```python
# drop rows that have NaN in any of the main features we engineered
features = ["return", "log_return", "ma_5", "ma_20", "vol_20"]
df_clean = df.dropna(subset=features)

df_clean.info()
```

```python
# weekly OHLCV (Monday–Friday grouped to one candle; Pandas default week ending Sunday)
weekly = df.resample("W").agg({
    "open": "first",
    "high": "max",
    "low":  "min",
    "close": "last",
    "volume": "sum"
})

weekly.head()
```

```python
weekly = weekly.dropna(subset=["open", "high", "low", "close"])
weekly.head()
weekly.info()
```

```python
plt.figure(figsize=(10, 4))
weekly["close"].plot()
plt.title("IKCO - Weekly Closing Price")
plt.xlabel("Week")
plt.ylabel("Price")
plt.show()
```
